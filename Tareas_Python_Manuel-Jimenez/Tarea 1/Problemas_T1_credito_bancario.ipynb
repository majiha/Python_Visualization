## Problem 1: Banking Case Study: Risk Management

* Analyze and plot the risk of bad loans by ages.
* Which is safer, leaving money to the young or the old people?

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('loansRisk.csv', sep=',', na_values=': ')
df
list(df.columns.values)
```

```python
import seaborn as sns
sns.lineplot(data = df)
```

```python
sns.barplot(x="Age group", y="Bad Loans", data=df)
```

Podemos intentar mostrar la proporcion de bad loans sobre el total


```python
df['Bad_Total'] = df['Bad Loans'] / df['Total Loans']
df.head
sns.barplot(x="Age group", y="Bad_Total", data=df)
#sns.lineplot(data = df, x='Age group', y='Bad_Total')
```

En este gráfico vemos claramente como el mayor porecentaje de impago (Bad loans) lo tienen las personas más jovenes (del primer grupo) con una amplia mayoría. A pesar de que en el gráfico de más arriba se ve que la muestra es más pequeña ya que son los que menos préstamos reciben.
